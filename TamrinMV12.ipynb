```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
image = cv2.imread('hidden.png', cv2.IMREAD_GRAYSCALE)

# Apply CLAHE (Contrast Limited Adaptive Histogram Equalization)
clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8, 8))
enhanced_image = clahe.apply(image)

# Show original and enhanced image side-by-side
plt.figure(figsize=[12, 6])

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(enhanced_image, cmap='gray')
plt.title("Enhanced Contrast with CLAHE")
plt.axis('off')

plt.tight_layout()
plt.show()

# Optionally save the enhanced image
cv2.imwrite('enhanced_hidden.png', enhanced_image)
```
