```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')
data_path = "/content/drive/MyDrive/Colab Notebooks/class.vision-master/images"
image = cv2.imread(f'{data_path}/akhavan.jpg')
```

```python
# ایجاد یک تصویر سفید 500x500
img = np.ones((500, 500, 3), dtype=np.uint8) * 255

# رسم بدنه‌ی خانه (مستطیل)
start_point = (150, 250)
end_point = (350, 450)
cv2.rectangle(img, start_point, end_point, color=(0, 0, 0), thickness=2)

# رسم سقف خانه (مثلث)
triangle_points = np.array([[150, 250], [350, 250], [250, 150]], np.int32)
triangle_points = triangle_points.reshape((-1, 1, 2))
cv2.polylines(img, [triangle_points], isClosed=True, color=(0, 0, 0), thickness=2)

# نمایش تصویر
cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
