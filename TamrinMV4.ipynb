```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')
data_path = "/content/drive/MyDrive/Colab Notebooks/class.vision-master/images"
img_original = cv2.imread(f'{data_path}/Weeki_Wachee_spring_10079u.jpg')
```

```python
import cv2
import numpy as np

# Define the gamma value
gamma = 0.4

# Apply linear brightness adjustment (alpha = 1.3, beta = 40)
output_linear = cv2.convertScaleAbs(img_original, alpha=1.3, beta=40)

# Create the lookup table for gamma correction
lookUpTable = np.empty((1, 256), np.uint8)

for i in range(256):
    lookUpTable[0, i] = np.clip(pow(i / 255.0, gamma) * 255.0, 0, 255)

# Apply gamma correction
output_gammacorrected = cv2.LUT(img_original, lookUpTable)

# Display the images (if using Jupyter Notebook or Google Colab, use cv2_imshow)
from google.colab.patches import cv2_imshow


print("original")
cv2_imshow(img_original)            # Original image
print("output_linear")
cv2_imshow(output_linear)           # Linear adjusted image
print("output_gammacorrected")
cv2_imshow(output_gammacorrected)   # Gamma corrected image

# If running locally, use cv2.imshow() instead:
# cv2.imshow('Linear Adjusted Image', output_linear)
# cv2.imshow('Gamma Corrected Image', output_gammacorrected)

```
